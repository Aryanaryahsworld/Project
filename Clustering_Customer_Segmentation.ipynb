```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("Mall_Customers.csv")
data
```

```python
# extracting the features
features = data.iloc[:,[3,4]].values           #indicates Age and Annual income
features
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(features[:,0],features[:,1])
plt.title("Customer Segmentation Plot")
plt.xlabel("Estimated Salary")
plt.ylabel("Spending Score")
```

```python
# lets extract the features in two different columns
features[:,0]
#features[:,1]
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters = 5)
```

```python
pred_cluster = kmeans.fit_predict(features)
```

```python
features
```

```python
len(features)
```

```python
features[pred_cluster == 0]
len(features[pred_cluster == 0])
```

```python
len(features[pred_cluster == 1])
len(features[pred_cluster == 2])
len(features[pred_cluster == 3])
len(features[pred_cluster == 4])
len(features[pred_cluster == 5])
```

```python
plt.scatter(features[pred_cluster == 0,0],features[pred_cluster == 0,1], c = 'blue', label = 'Cluster 1')
plt.scatter(features[pred_cluster == 1,0],features[pred_cluster == 1,1], c = 'red', label = 'Cluster 2')
plt.scatter(features[pred_cluster == 2,0],features[pred_cluster == 2,1], c = 'green', label = 'Cluster 3')
plt.scatter(features[pred_cluster == 3,0],features[pred_cluster == 3,1], c = 'cyan', label = 'Cluster 4')
plt.scatter(features[pred_cluster == 4,0],features[pred_cluster == 4,1], c = 'magenta', label = 'Cluster 5')
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1], c = 'yellow', label = 'Centroids')
plt.title('Clusters of Customers')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.show()
```
